# 1.How many are there in each Team and the percentage splitting with respect to the total employees.


```python
df['Team'].value_counts()
```

```python
plt.figure(figsize=(25,8))
plt.xticks(rotation="vertical")
sns.countplot(x='Team',data=df,hue='Team',palette='plasma')
```

```python
no_of_team=30
```

```python
no_of_employee=458
```

```python
team_series=pd.Series(df['Team'].value_counts())
percentage=(team_series/no_of_employee)*100
result=pd.concat([team_series,percentage.round(2)],axis=1)
result.columns=['Number of employee','Percentage']
result
```

Data Story : From this data we can analyse that there are 30 teams and majority of the team possess nearly 15 employees.The countplot analysis based on the employee teams reveals that the 'New Orleans Pelicans' team has the highest number of employees, totaling 19, closely followed by the 'Memphis Grizzlies' team, which comprises 18 employees.

# 5. Find if there is any correlation between age and salary, and represent it visually.


```python
df2=df[['Age','Salary']]
df2
```

```python
corr=df2.corr()
corr
```

```python
Data Story: There exists a positive correlation between Age and Salary where the values lie between 0 and 1
However, it is evident from the results that the correlation is not very high.
```

```python
sns.heatmap(corr, annot=True, cmap="viridis")
plt.title("Correlation Matrix between Age and Salary of Employee Dataset")
plt.show()
```
